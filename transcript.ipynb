```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('academic_results.csv')
df.head()
```

```python
print(df.columns)
```

```python
phase_1 = df[ df['phase'] == 1]

print(phase_1.head())
```

```python
score_distribution_phase1= phase_1['score'].value_counts().sort_index()
print("=== Score Distribution (Phase 1) ===")
print(score_distribution_phase1)

score_catergory_phase1 =phase_1.groupby('catergory')['score'].mean().sort_values(ascending=False)
print("\n=== Average Score by Category (Phase 1) ===")
print(score_catergory_phase1)
```

```python
phase_2 = df[df['phase'] == 2]
print(phase_2)
```

```python
score_distribution_phase2 = phase_2['score'].value_counts().sort_index()
print("=== Score Distribution (Phase 2) ===")
print(score_distribution_phase2)

score_catergory_phase2 = phase_2.groupby('catergory')['score'].mean().sort_values(ascending=False)
print("\n=== Average Score by Category (Phase 2) ===")
print(score_catergory_phase2)
```

```python
phase_comparison = pd.concat([score_catergory_phase1,score_catergory_phase2], axis=1)

phase_comparison.columns = ['Phase 1 Avg', 'Phase 2 Avg']

phase_comparison['Delta'] = (phase_comparison['Phase 2 Avg'] - phase_comparison['Phase 1 Avg'])

phase_comparison = phase_comparison.fillna('—')

print(phase_comparison)
```

```python
phase_3 = df[df['phase']== 3]

print(phase_3)
```

```python
score_distribution_phase3 = phase_3['score'].value_counts().sort_index()
print("\nScore Distribution (Phase 3)")
print("-" * 30)
print(score_distribution_phase3)

print("\nAverage Score by Category (Phase 3)")
print("-" * 35)
score_catergory_phase3= phase_3.groupby('catergory')['score'].mean().sort_values(ascending=False)
print(score_catergory_phase3)
```

```python
phase_comparison = pd.concat([score_catergory_phase1,score_catergory_phase2, score_catergory_phase3], axis=1)

phase_comparison.columns = ['Phase 1 Avg', 'Phase 2 Avg', 'Phase 3 Avg']

phase_comparison['Trend'] = (phase_comparison['Phase 3 Avg'] - phase_comparison['Phase 1 Avg'])

phase_comparison = phase_comparison.fillna('—')

print(phase_comparison)
```

```python
# Data Overview
print("=" * 60)
print("ACADEMIC RESULTS SUMMARY")
print("=" * 60)
print(f"\nTotal Courses: {len(df)}")
print(f"Total Credits: {df['credits'].sum()}")

# Performance Statistics by Category
print("\n" + "=" * 60)
print("PERFORMANCE STATISTICS BY CATEGORY")
print("=" * 60)

category_stats = df.groupby('catergory')['score'].agg(['count', 'mean', 'min', 'max']).round(2)
category_stats.columns = ['Courses', 'Avg Score', 'Min', 'Max']
print(category_stats.sort_values('Avg Score', ascending=False))


# Score Distribution Analysis
print("\n" + "=" * 60)
print("SCORE DISTRIBUTION ANALYSIS")
print("=" * 60)

# Score ranges
ranges = [(0, 9, "Below 10"), (10, 13, "10-13"), (14, 17, "14-17"), (18, 20, "18-20")]
for low, high, label in ranges:
    count = len(df[(df['score'] >= low) & (df['score'] <= high)])
    pct = (count / len(df)) * 100
    print(f"  {label}: {count} courses ({pct:.1f}%)")

```
